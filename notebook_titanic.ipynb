```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(18,6))
df.Survived.value_counts(normalize=True).plot(kind="bar")
plt.title("Survived")
```

```python
fig = plt.figure(figsize=(18,6))
df.Pclass.value_counts(normalize=True).plot(kind="bar")
plt.title("Passenger class")
```

```python
fig = plt.figure(figsize=(18,6))
for x in [1,2,3]:
    df.Age[df.Pclass == x].plot(kind="kde")
```

```python
df.Embarked.value_counts(normalize=True).plot(kind="bar")
```

```python
fig = plt.figure(figsize=(30,15))
plt.subplot2grid((3,4), (0,0))
df.Survived.value_counts(normalize=True).plot(kind="bar")
plt.title("Global Survival Rate")

plt.subplot2grid((3,4), (0,1))
df.Survived[df.Sex == "male"].value_counts(normalize=True).plot(kind="bar")
plt.title("Male Survival Rate")

plt.subplot2grid((3,4), (0,2))
df.Survived[df.Sex == "female"].value_counts(normalize=True).plot(kind="bar")
plt.title("Female Survival Rate")
```

```python
fig = plt.figure(figsize=(30,15))
plt.subplot2grid((3,4), (0,0))
df.Survived[df.Pclass == 1].value_counts(normalize=True).plot(kind="bar")
plt.title("Rich Survival Rate")

fig = plt.figure(figsize=(30,15))
plt.subplot2grid((3,4), (0,1))
df.Survived[df.Pclass == 3].value_counts(normalize=True).plot(kind="bar")
plt.title("Poor Survival Rate")
```

```python
fig = plt.figure(figsize=(30,15))
plt.subplot2grid((3,4), (0,0))
df.Survived[(df.Pclass == 1) & (df.Sex == "male")].value_counts(normalize=True).plot(kind="bar")
plt.title("Rich Male Survival Rate")

plt.subplot2grid((3,4), (0,1))
df.Survived[(df.Pclass == 3) & (df.Sex == "male")].value_counts(normalize=True).plot(kind="bar")
plt.title("Poor Male Survival Rate")

plt.subplot2grid((3,4), (1,0))
df.Survived[(df.Pclass == 1) & (df.Sex == "female")].value_counts(normalize=True).plot(kind="bar")
plt.title("Rich Female Survival Rate")

plt.subplot2grid((3,4), (1,1))
df.Survived[(df.Pclass == 3) & (df.Sex == "female")].value_counts(normalize=True).plot(kind="bar")
plt.title("Poor Female Survival Rate")


```
